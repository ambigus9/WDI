# Filtro de frecuencia indicador en serie temporal

Por medio de la siguiente función buscamos derminar la cantidad de registros para cada indicador en cada continente:

- Cargamos el CSV del continente, agrupamos en base a cuatro columnas: Nombre del Indicador y Nombre del Pais, Código del indicador y Código del Pais
- Utilizamos el método count() para efectuar un conteo de las repeticiones de los indicadores registrados en la serie temporal

```python
def filtro_frecuencia_indicador(datos_continente):
    temp = datos_continente.drop('Unnamed: 0', 1).groupby(['IndicatorName','CountryName','IndicatorCode','CountryCode'])
    agrupado = temp.count()
    return agrupado
```

Guardarmos en un CSV para cada caso

```python
filtro_frecuencia_indicador(pd.read_csv("suramerica.csv")).to_csv('suramerica2.csv')
filtro_frecuencia_indicador(pd.read_csv("europa.csv")).to_csv('europa2.csv')
filtro_frecuencia_indicador(pd.read_csv("este.csv")).to_csv('este2.csv')
filtro_frecuencia_indicador(pd.read_csv("africa.csv")).to_csv('africa2.csv')
filtro_frecuencia_indicador(pd.read_csv("sudafrica.csv")).to_csv('sudafrica2.csv')
```

# Media de las Frecuencias

Por medio de la siguiente función calculamos para un conjunto de indicadores en específico la Media de la frencuencia de los indicadores

```python
def media_frencuencias(continente,datos_continente,indicadores):
    filtro = []
    temporal = list()
    grupo = list()
    output = {}
    
    for k in range(len(datos_continente)):
        cn = datos_continente['CountryName'][k]
        indicator_code = datos_continente['IndicatorCode'][k]

        if cn in continente and indicator_code in indicadores:
            year = datos_continente['Year'][k]

            if cn not in output.keys():
                output[cn] = []
            output[cn].append(year)
       
    for p in continente:
        if p not in output.keys():
            temporal.append(np.array([p,0]))

            
    for o in output:
        temporal.append(np.array([o,np.mean(output.get(o))]))

    temporal = sorted(temporal,key=lambda x:x[0])  

    for i in temporal:
        filtro.append(i[1])
    grupo = filtro
    return grupo
```

# Tabla 1

Definimos:

- X indicadores con los cuales entrenaremos el modelo buscando predecir un indicador Y
- Año inicial
- Rango de años (Si es negativo anteriores, si es positivo posteriores)

```python
x_indicators = np.array(['EG.ELC.ACCS.RU.ZS','EG.NSF.ACCS.RU.ZS','SP.ADO.TFRT'])
y_indicator = np.array(['SE.PRM.TENR'])
year_init = 2012
year_range = -5
```

```python
def tabla1 (continente,datos_continente,x_indicators,y_indicator,year_init,year_range):
    
    datos = datos_continente
    paises = continente
    
    #Creamos lista de años a tratar en base al year_range y year_init
    years = list()
    if year_range >0:
        for i in range(year_range):
            years.append(year_init+i)
    if year_range <0:
        for i in range(abs(year_range)):
            years.append(year_init-i)
            
    #Creamos una lista temporal que contiene todos los valores de los indicadores en 0          
    indicadores = list()
    for i in x_indicators: indicadores.append(i)
    for i in y_indicator: indicadores.append(i)
    temporal = list()

    for i in paises:
        for j in indicadores:
            for k in years:
                temporal.append([i,k,j,0.])
                
    #Agregamos valores de los indicadores a nuestra lista temporal              
    indicadores2 = list()
    indicadores2.append('Pais')
    indicadores2.append('Year')

    for k in range(len(datos)):
        cn = datos['CountryName'][k]
        indicator_code = datos['IndicatorCode'][k]
        indicator_name = datos['IndicatorName'][k]
        y = datos['Year'][k]
        v = datos['Value'][k]

        for i in range(len(temporal)):
            if cn == temporal[i][0] and y == temporal[i][1] and indicator_code == temporal[i][2]:
                temporal[i][3] = v

        for j in indicadores:
            if indicator_code == j and indicator_name not in indicadores2:
                indicadores2.append(indicator_name)

        
    #Consolidamos la tabla 1 en base a la lista temporal   
    valores = list()
    paises2 = list()
    years2 = list()
    matriz = []
    matriz2 = []

    for i in paises:
        for j in years: paises2.append(i)
    matriz2.append(paises2)        

    for i in temporal: years2.append(i[1])
    matriz2.append(years2)

    for i in range(len(indicadores)):
        for k in range(len(temporal)):
            if indicadores[i] == temporal[k][2]:
                valores.append(temporal[k][3])
        matriz.append(valores)
        matriz2.append(valores)
        valores = list()

    matriz = zip(*matriz)
    matriz = np.array(matriz)
    matriz2 = zip(*matriz2)
    
    return pd.DataFrame(matriz2,columns=indicadores2)
```

```python
tabla1(suramerica,pd.read_csv("suramerica.csv"),x_indicators,y_indicator,year_init,year_range)
```

```python
from sklearn.datasets import *
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
%matplotlib inline

def plot_2Ddata_with_boundary(predict,X,y):
    n = 200
    mins,maxs = np.min(X,axis=0), np.max(X,axis=0)
    mins -= np.abs(mins)*.2
    maxs += np.abs(maxs)*.2
    d0 = np.linspace(mins[0], maxs[0],n)
    d1 = np.linspace(mins[1], maxs[1],n)
    gd0,gd1 = np.meshgrid(d0,d1)
    D = np.hstack((gd0.reshape(-1,1), gd1.reshape(-1,1)))
    p = (predict(D)*1.).reshape((n,n))
    plt.contourf(gd0,gd1,p, levels=[-0.1,0.5], alpha=0.5, cmap=plt.cm.Greys)
    plt.scatter(X[y==0][:,0], X[y==0][:,1], c="blue")
    plt.scatter(X[y==1][:,0], X[y==1][:,1], c="red")

def crossval_estimator(estimator, X,y, n_folds=5):
    from sklearn.cross_validation import KFold
    sctr, scts = [],[]
    for itr, its in KFold(len(X), n_folds):
        X_train = X[itr]
        y_train = y[itr]
        X_test  = X[its]
        y_test  = y[its]

        estimator.fit(X_train,y_train)
        
        sctr.append(estimator.score(X_train,y_train))
        scts.append(estimator.score(X_test,y_test))
    return np.mean(sctr), np.std(sctr), np.mean(scts), np.std(scts)

import urllib, inspect
src3 = urllib.quote_plus(inspect.getsource(crossval_estimator))

from sklearn.tree import DecisionTreeClassifier

def crossval_estimators(estimators,X,y):
    meanstr, stdstr = [], []
    meansts, stdsts = [], []
    for estimator in estimators:
        mstr, sstr, msts, ssts = crossval_estimator(estimator, X, y, 5)
        meanstr.append(mstr)
        stdstr.append(sstr)
        meansts.append(msts)
        stdsts.append(ssts)

    return [np.array(i) for i in meanstr, meansts, stdstr, stdsts]

X,y = matriz[:,:(len(indicadores)-1)], np.asarray(matriz[:,(len(indicadores)-1)], dtype="|S6")

rngs = range(1,10,1)
estimators = [DecisionTreeClassifier(max_depth=i) for i in rngs]

def plot_cv_estimators(estimators, rngs, X,y):
    meanstr, meansts, stdstr, stdsts = crossval_estimators(estimators,X,y)

    plt.grid()
    plt.plot(rngs, meanstr, "o-", color="red", label="train")
    plt.fill_between(rngs, meanstr-stdstr, meanstr+stdstr, color="red", alpha=.2)

    plt.plot(rngs, meansts, "o-", color="green", label="test")
    plt.fill_between(rngs, meansts-stdsts, meansts+stdsts, color="green", alpha=.2)
    plt.xlabel("estimators")
    plt.ylabel("score")
    plt.legend(loc="best")
    plt.title("validation curve")
    
plot_cv_estimators(estimators, rngs, X,y)
```
