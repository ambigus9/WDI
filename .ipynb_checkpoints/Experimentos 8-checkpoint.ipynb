```python
indicadores = np.array(['EG.ELC.ACCS.RU.ZS','EG.NSF.ACCS.RU.ZS'])
years = np.array([2011,2012])
```

```python
start_time = time.time()
temporal = list()

for i in paises:
    for j in indicadores:
        for k in years:
            temporal.append(np.array([i,k,j,0]))
            
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
temporal = sorted(temporal,key=lambda x:x[0]) 

filtro1 = list()
filtro2 = list()
filtro3 = list()
filtro4 = list()

for i in temporal: filtro1.append(i[0])
for i in temporal: filtro2.append(i[1])
for i in temporal: filtro3.append(i[2])
for i in temporal: filtro4.append(i[3])
    
pd.DataFrame(
    {'Paises': filtro1,
     'Año': filtro2,
     'Indicador': filtro3,
     'Valor': filtro4,
    }, columns=['Paises','Año','Indicador','Valor'])
```
