```python
import numpy as np
import pandas as pd
paises = np.array(['Antigua and Barbuda','Argentina','Aruba','Bahamas, The','Barbados','Belize','Bolivia','Brazil','Cayman Islands','Chile','Colombia','Costa Rica','Cuba','Curacao','Dominica','Dominican Republic','Ecuador','El Salvador','Grenada','Guatemala','Guyana','Haiti','Honduras','Jamaica','Mexico','Nicaragua','Panama','Paraguay','Peru','Puerto Rico','Sint Maarten (Dutch part)','St. Kitts and Nevis','St. Vincent and the Grenadines','St. Lucia','St. Martin (French part)','Suriname','Trinidad and Tobago','Turks and Caicos Islands','Uruguay','Venezuela, RB','Virgin Islands (U.S.)'])
datos = pd.read_csv("avance2.csv")
datos
```

```python
a = datos['CountryName'].values.tolist()
c = np.array(['Antigua and Barbuda','Argentina','Aruba','Bahamas, The','Barbados','Belize','Bolivia','Brazil','Cayman Islands','Chile','Colombia','Costa Rica','Cuba','Curacao','Dominica','Dominican Republic','Ecuador','El Salvador','Grenada','Guatemala','Guyana','Haiti','Honduras','Jamaica','Mexico','Nicaragua','Panama','Paraguay','Peru','Puerto Rico','Sint Maarten (Dutch part)','St. Kitts and Nevis','St. Vincent and the Grenadines','St. Lucia','St. Martin (French part)','Suriname','Trinidad and Tobago','Turks and Caicos Islands','Uruguay','Venezuela, RB','Virgin Islands (U.S.)'])
pais = list()
for i in c:
    offset = 0;
    for j in range(a.count(i)):
        pais.append(a.index(i,offset))
        offset = pais[-1]+1
pais
```

```python
indicadores_ayuda = np.array(['DT.DOD.MDRI.CD','BX.GRT.EXTA.CD.WD','DT.DIS.IDAG.CD','DC.DAC.AUSL.CD'])
```

```python
b = datos['IndicatorCode'].values.tolist()
d = indicadores_ayuda
indicadores = list()
for i in d:
    offset = 0;
    for j in range(b.count(i)):
        indicadores.append(b.index(i,offset))
        offset = indicadores[-1]+1
indicadores
```

```python
media = list()
mejora = list()
flag=0
flag2=0

for i in c:
    if flag2==0:
        for j in pais:
            flag2=1
            for k in indicadores:
                if j==k:
                    flag=1
                    media.append(datos['Year'][j])
                    #print datos['CountryName'][j],datos['IndicatorName'][j] , media
                    #mejora.append(np.array([datos['CountryName'][j],datos['IndicatorName'][j],media]))
            if flag==1:
                mejora.append([datos['CountryName'][j],datos['IndicatorName'][j],np.mean(media)])
                del media[:]
                flag=0
                flag2=0
pd.DataFrame(mejora)
```
