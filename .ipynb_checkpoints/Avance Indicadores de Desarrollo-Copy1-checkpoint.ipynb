# Filtro Indicadores

```python
def filtro_indicadores(continente,datos_continente,indicadores):
    indices = list()
    
    for k in range(len(datos_continente)):
        cn = datos_continente['CountryName'][k]
        indicator_code = datos_continente['IndicatorCode'][k]

        if cn in continente and indicator_code in indicadores:
            indices.append(k)
    
    return datos_continente.reindex([indices])
```

# Gráfica Filtro Indicadores

```python
def grafica_filtro_indicadores(continente,datos_continente,indicadores):
    suramerica_indicadores = filtro_indicadores(continente,datos_continente,indicadores)
    data = [
        go.Heatmap(
            z=suramerica_indicadores['Year'].values.tolist(),
            x=suramerica_indicadores['IndicatorName'].values.tolist(),
            y=suramerica_indicadores['CountryName'].values.tolist(),
            colorscale='Viridis'
        )
    ]
    return py.iplot(data, filename='Agricultura - Suramerica')
```

# Cross Validation

Aplicamos un Cross validation a la tabla de datos normalizada

```python
from sklearn.datasets import *
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
%matplotlib inline

def crossval_estimator(estimator, X,y, n_folds=15):
    from sklearn.cross_validation import KFold
    sctr, scts = [],[]
    for itr, its in KFold(len(X), n_folds):
        
        X_train = X[itr]
        y_train = y[itr]
        X_test  = X[its]
        y_test  = y[its]

        estimator.fit(X_train,y_train)
        
        sctr.append(estimator.score(X_train,y_train))
        scts.append(estimator.score(X_test,y_test))
    
    return np.mean(sctr), np.std(sctr), np.mean(scts), np.std(scts)

from sklearn.tree import DecisionTreeClassifier

def crossval_estimators(estimators,X,y):
    meanstr, stdstr = [], []
    meansts, stdsts = [], []
    for estimator in estimators:
        mstr, sstr, msts, ssts = crossval_estimator(estimator, X, y)
        meanstr.append(mstr)
        stdstr.append(sstr)
        meansts.append(msts)
        stdsts.append(ssts)

    return [np.array(i) for i in meanstr, meansts, stdstr, stdsts]

X,y = Xr[:,8:9], np.asarray(Xr[:,8], dtype="|S6")


rngs = range(1,15,1)
estimators = [DecisionTreeClassifier(max_depth=i) for i in rngs]

def plot_cv_estimators(estimators, rngs, X,y):
    meanstr, meansts, stdstr, stdsts = crossval_estimators(estimators,X,y)

    print meanstr[-1]
    print meansts[-1]
    
    plt.grid()
    plt.plot(rngs, meanstr, "o-", color="red", label="train")
    plt.fill_between(rngs, meanstr-stdstr, meanstr+stdstr, color="red", alpha=.2)

    plt.plot(rngs, meansts, "o-", color="green", label="test")
    plt.fill_between(rngs, meansts-stdsts, meansts+stdsts, color="green", alpha=.2)
    plt.xlabel("estimators")
    plt.ylabel("score")
    plt.legend(loc="best")
    plt.title("validation curve")
    
plot_cv_estimators(estimators, rngs, X,y)
```
